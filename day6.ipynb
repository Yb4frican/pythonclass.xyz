```python
# functions with more than one or multiple parameters
def person(name,gender,height,weight,age):
    print(f'''Name:{name}
    Gender:{gender}
    Height:{height}
    Weight:{weight}
    Age:{age}
    ''')
    if age >=18:
        print('congratulations you are eligible to vote')
    else:
        print('oh!!!!! you are not eligible to vote')
person('Simeon', 'male',1.75,69,30)
    
```

```python
#4.
def isprime():
    num = int(input('enter the first number:'))
    if num>1:
        for i in range(2,num):
            if(num % 1)==0:
                print(num,'is not a prime number')
                print(i,'times',num//i, 'is,num') #note; the two slash// is called interger division symbol
                break #note; this is used to terminate a program.
            else:
                print(num,'is a prime number')
    else:
        print(num, 'is not a prime number')
isprime()
```
